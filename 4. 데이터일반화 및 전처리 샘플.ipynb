```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 간단한 난수 생성
x, y = make_regression(n_samples=100, n_features=1, noise=15.3) # standard deviation
```

```python
print(x)
```

```python
print(y)
```

```python
plt.scatter(x, y)
```

```python
plt.plot(x)
plt.show()
```

```python
plt.plot(y)
plt.show()
```

```python
# 데이터 분포 확인하기, x
plt.hist(x, bins=50)
```

```python
# 데이터 분포 확인하기, y
plt.hist(y, bins=50)
```

Min-Max normalization


```python
mM_x = []
for i in range(len(y)):
    mM_x.append((y[i]-min(y))/(max(y)-min(y)))

print(mM_x)
```

```python
plt.scatter(y, mM_x)
```

```python
plt.hist(mM_x, bins=50)
```

Z-score normalization


```python
# Z-score normalization
Zsc_x = []
for i in range(len(y)):
    Zsc_x.append((y[i]-np.mean(y))/(np.std(y)))

print(Zsc_x)
```

```python
plt.scatter(y, Zsc_x)
```

```python
plt.hist(Zsc_x, bins=50)
```

Max normalization


```python
# Max normalization
Mx_x = []
for i in range(len(y)):
    Mx_x.append((y[i])/(max(y)))
```

```python
print(Mx_x)
```

```python
plt.scatter(y, Mx_x)
```

```python
plt.hist(Mx_x, bins=50)
```
